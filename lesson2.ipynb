```python
# 3
t = np.linspace(0, 10, 51)
from math import cos
f = np.frompyfunc(cos, 1, 1)(t)
plt.plot(t, f, color='green')
plt.title('График f(x)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

```python
# 4
x = np.linspace(-3, 3, 51)
y1 = np.vectorize(lambda e: e**2)(x)
y2 = np.vectorize(lambda e: 2 * e + 0.5)(x)
y3 = np.vectorize(lambda e: -3 * e - 1.5)(x)
from math import sin
y4 = np.frompyfunc(sin, 1, 1)(x)
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(6.0)
fig.set_figwidth(8.0)
fig.set_constrained_layout_pads(w_pad=0.3, h_pad=0.3)
ax[0][0].plot(x, y1)
ax[0][0].set_xlim(-5, 5)
ax[0][0].set_title('График y1')
ax[0][1].plot(x, y2)
ax[0][1].set_title('График y2')
ax[1][0].plot(x, y3)
ax[1][0].set_title('График y3')
ax[1][1].plot(x, y4)
ax[1][1].set_title('График y4')
plt.show()
```

```python
# 5
import pandas as pd
# downloaded manually, because there was authorisation
ccard = pd.read_csv('creditcard.csv')
plt.style.use('fivethirtyeight')
```

```python
pd.value_counts(ccard['Class']).plot(kind='bar')
```

```python
pd.value_counts(ccard['Class']).plot(kind='bar', logy=True)
```

```python
class0 = ccard.loc[ccard.loc[:, 'Class'] == 0, 'V1']
class1 = ccard.loc[ccard.loc[:, 'Class'] == 1, 'V1']
plt.hist(class0, 20, alpha=0.5, density=True, color='gray', label='Class 0')
plt.hist(class1, 20, alpha=0.5, density=True, color='red', label='Class 1')
plt.legend()
plt.xlabel('Class')
plt.show()
```
