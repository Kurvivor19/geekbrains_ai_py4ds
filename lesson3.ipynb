```python
# 1
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target, columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
linear_r2 = r2_score(y_test, y_pred)
```

```python
# 2
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred2 = model.predict(X_test)
```

```python
forest_r2 = r2_score(y_test, y_pred2)
print('Linear regression prediction R2 score: %lf\nRandom forest model R2 score: %lf' % (linear_r2, forest_r2))
```

```python
# согласно полученным значениям, модель случайного леса дает лучшее приближение чем линейная регрессия
```

```python
# 3
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#plt.plot(model.feature_importances_, kind='bar')

#
fimp = pd.Series(model.feature_importances_)
fplot = fimp.plot(kind='bar')
fplot.set_xticklabels(X.columns.values)
plt.show()
```

```python
weight_sum = sum(model.feature_importances_)
secondbest_ind, best_ind = np.argsort(model.feature_importances_)[-2:]
print('Sum of argument weights is %f\nMost important feature is %s with weight %f\nSecond in importance feature is %s with weight %f' % (weight_sum, X.columns.values[best_ind], model.feature_importances_[best_ind],  X.columns.values[secondbest_ind], model.feature_importances_[secondbest_ind]))
```

```python
# 4
from sklearn.model_selection import GridSearchCV
df = pd.read_csv('creditcard.csv')
```

```python
df.info()
# всу столбцы не-нулевые и числовые
```

```python
pd.value_counts(df.loc[:, 'Class'], normalize=True)
# класс 0 занимет чуть больше 99,8%
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
X = df.copy()
del X['Class']
X.info()
```

```python
y = pd.Series(df.loc[:, 'Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
parameters = [
    {'n_estimators': [10, 15],
     'max_features': np.arange(3, 5),
     'max_depth': np.arange(4, 7)}
]
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
y_pred_proba = grid_search.predict_proba(X_test)
y_pred_proba = y_pred_proba[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_test_score = roc_auc_score(y_test, y_pred_proba)
```

```python
y_pred_proba_train = grid_search.predict_proba(X_train)[:, 1]
roc_auc_train_score = roc_auc_score(y_train, y_pred_proba_train)
print('Areas under curve:\nOn training data: %lf\nOn test data: %lf' % (roc_auc_train_score, roc_auc_test_score))
```
